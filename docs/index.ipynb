# What is specsuite?

```specsuite``` is a Python data reduction toolbox built for ground-based long-slit spectrographs. It is capable of...

- Loading FITS data into arrays
- Applying a flatfield correction
- De-warping images using arclamp exposures
- Extracting sky backgrounds
- Interactively creating a wavelength calibration

We designed ```specsuite``` to be as friendly to new users as possible.
